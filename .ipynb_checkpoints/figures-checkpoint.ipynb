```python
datasets = ['normal', 'nopbs', 'f2', 'y10', '']  # Add your dataset names
bands = ['HF-2', 'HF-1', 'MF-2', 'MF-1', 'LF-2', 'LF-1']

# Extract mean and std for r bias for each dataset
data_dict = {}
for dataset in datasets:
    data_dict[dataset] = {
        'pos_means': [],
        'pos_stds': [],
        'neg_means': [],
        'neg_stds': []
    }
    # Fill with your actual data from bias_dict_[dataset]

# Create grouped bar chart
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

x = np.arange(len(bands))
width = 0.15
colors = ['red', 'blue', 'green', 'orange', 'purple']

# +2% bias plot
for i, dataset in enumerate(datasets):
    ax1.bar(x + i*width, data_dict[dataset]['pos_means'], width, 
           yerr=data_dict[dataset]['pos_stds'], label=dataset, 
           color=colors[i], alpha=0.7)

ax1.set_xlabel('Frequency Bands')
ax1.set_ylabel('r bias')
ax1.set_title('+2% Bias')
ax1.set_xticks(x + width*2)
ax1.set_xticklabels(bands)
ax1.legend()
ax1.grid(alpha=0.3)

# -2% bias plot
for i, dataset in enumerate(datasets):
    ax2.bar(x + i*width, data_dict[dataset]['neg_means'], width, 
           yerr=data_dict[dataset]['neg_stds'], label=dataset, 
           color=colors[i], alpha=0.7)

ax2.set_xlabel('Frequency Bands')
ax2.set_ylabel('r bias')
ax2.set_title('-2% Bias')
ax2.set_xticks(x + width*2)
ax2.set_xticklabels(bands)
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
