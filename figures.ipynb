```python
import matplotlib.pyplot as plt
import h5py
import numpy as np
import pandas as pd
import phase2_common as ph2
from s4bb.spectra import XSpec, CalcSpec_namaster
```

```python
with h5py.File('/global/homes/v/vhoffman/reproduce_colin_results/spectra/phase2_spec_comb_f1_y20_n3_full_withpbs_000_100.h5', 'r') as f:
    spec = XSpec.from_hdf5(f)

ell_centers = 0.5 * (spec.bins[0] + spec.bins[1])

ps_mean = np.mean(spec.spec, axis=2)

spec_names = spec.str()

plt.plot(ell_centers, ps_mean[9], color='red')
plt.xlabel(r'$\ell$')
plt.ylabel(r'$C_\ell$')
plt.title(f'{spec_names[9]}')
plt.grid()
plt.show()
```
